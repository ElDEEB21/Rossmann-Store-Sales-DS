```python
import pandas as pd
import joblib
import matplotlib.pyplot as plt

# Load the trained XGBRegressor model
model = joblib.load('D:\\model\\XGBRegressor_best_model.pkl')

# Extract the feature importance scores
importance = model.feature_importances_

# Define your feature names
features = ['CompetitionDistance', 'SalesPerCustomer', 'SalesPerDistance', 'CompetitionOpenSince', 'Promo2Since',
            'Store', 'DayOfWeek', 'Customers', 'Open', 'Promo', 'SchoolHoliday', 'Promo2', 'Year', 'Month', 
            'Day', 'WeekOfYear', 'StateHoliday', 'StoreType', 'PromoInterval', 'Assortmenta', 'Assortmentb']

# Create a DataFrame to display feature importance
importance_df = pd.DataFrame({
    'Feature': features, 
    'Weight': importance
})

# Sort the features by importance and get the top 5 (or top N)
top_importance_df = importance_df.sort_values(by='Weight', ascending=False).head(7)

# Save the top feature importances to a CSV
top_importance_df.to_csv("top_weights.csv", index=False)

# Display the DataFrame
print(top_importance_df)

# Plotting top feature importances
plt.figure(figsize=(10, 6))
plt.barh(top_importance_df['Feature'], top_importance_df['Weight'], color='skyblue')
plt.xlabel('Importance')
plt.title('Top 7 Feature Importances from XGBRegressor')
plt.axvline(0, color='gray', linestyle='--')  # Add a vertical line at x=0
plt.grid(axis='x')
plt.tight_layout()

# Show the plot
plt.show()
```
